```python
# Challenge 3ü§∫ - 15 min
# Definir una clase madre que herede a sus clases hijas dos atributos de instancia. 
# Mostrar ambos atributos en pantalla desde los objetos de las clases hijas.

# Observaci√≥n: Editar las clases hijas para que no se inicialicen.


class AnimalesZoo:
    tipo= 'animal de zoo'
    #atributos de instancia
    def __init__(self, nombre, especie, edad):
        self.nombre=nombre
        self.especie=especie
        self.edad=edad
        #accion
    def tipo_de_animal (self):  
        print(self.nombre, self.especie,)
    def tiempo_en_el_zoo(self):
        print(f'lleva {self.edad} en el zoo')
#subclases
class Cocodrilo(AnimalesZoo):
    pielreptil='escamas'
    def cubierto_de(self):
        print(f'cubierto de {self.pielreptil}')
    
    
class avestruz (AnimalesZoo):
    pielave='plumas'
    def cubierto_de(self):
        print(f'cubierto de {self.pielave}')

animal1= Cocodrilo('zila', 'reptil','80 anhos')
animal2=avestruz('yiyo','ave','12 anhos')

animal1.tipo_de_animal()
animal2.cubierto_de()
```
